##Ex 2: Integrate $f(x)$ between limits, $\int_{a}^{b} f(x) \,dx$

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from scipy import *
import scipy.linalg as la
from IPython.display import Image
from scipy.integrate import quad, dblquad, tplquad

def f(x):
  a=1
  return a*x**2

x_lower=0; x_upper=np.pi
val,abserr=quad(f,x_lower,x_upper)

print("integral value =", val, ", absolute error =", abserr)
n=256
x=np.linspace(x_lower,x_upper,n,endpoint=True)
y=f(x)
plt.plot(x,y,'--', color='red',linewidth=4,alpha=1.00)
plt.fill_between(x,y,0,color='blue',alpha=.1)
plt.show()
```
